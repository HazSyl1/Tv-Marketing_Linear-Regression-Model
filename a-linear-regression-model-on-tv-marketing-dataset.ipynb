```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
d1=pd.read_csv("/kaggle/input/tvmarketingcsv/tvmarketing.csv")
```

```python
from matplotlib import pyplot as plt
plt.scatter(d1['TV'],d1["Sales"])
plt.title('TV vs Sales ')
plt.xlabel("Tv")
plt.ylabel('Sale')
plt.show()
```

```python
x=d1.iloc[:,:1].values
y=d1.iloc[:,1].values
```

```python
print(x[:5])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print("X:",x)

print("x_train:",x_train)
print("x_test",x_test)
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train) # Training the algorithm
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
y_pred=regressor.predict(x_test)
print(y_pred)
```

```python
df=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
print(df)
```

```python
from matplotlib import pyplot as plt
plt.plot(x_train,regressor.predict(x_train), color='red')
plt.scatter(d1['TV'],d1["Sales"])
plt.title('Tv vs Sales')
plt.xlabel("TV")
plt.ylabel('Sales ')
plt.show()
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2_s=r2_score(y_test,y_pred)
print(f"MAE:{mae} , MSE:{mse} , RMSE:{rmse}, R2:{r2}")
```

```python
train_score_lr = regressor.score(x_train, y_train)
test_score_lr = regressor.score(x_test, y_test)

print("The train score for lr model is: ", train_score_lr)
print("The test score for lr model is: ",test_score_lr)

```

```python
train_score_lr = regressor.score(x_train, y_train)
test_score_lr = regressor.score(x_test, y_test)

print("The train score for lr model is: ", train_score_lr)
print("The test score for lr model is: ",test_score_lr)

```

```python
actual_minus_predicted = sum((y_test - y_pred)**2)
actual_minus_actual_mean = sum((y_test - y_test.mean())**2)
r2 = 1 - actual_minus_predicted/actual_minus_actual_mean
print('R²:', r2)
```
